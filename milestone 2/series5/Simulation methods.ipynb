### Acceptance-Rejection
Implement the acceptance-rejection method for Example 1.1.15 in the lecture notes.
Generate 100 samples and plot the output in a histogram. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot, norm

def acceptreject(n):
    print('Discrete Acceptance Rejection')
    x = ar_randy(n)
    plt.hist(x)
    plt.show()

def ar_randy(n):
    p = [0.11, 0.12, 0.09, 0.08, 0.12, 0.10, 0.09, 0.09, 0.10, 0.10]
    j = 0
    x = np.zeros((n, 1))
    while j < n:
        y = discreterandu(1)
        u = np.random.rand(1)
        c = 1.2
        if u <= p[y[0]-1] / (c * 0.10):
            x[j, 0] = y[0]
            j += 1
    return x

def discreterandu(n):
    x = np.ceil(10 * np.random.rand(n))
    return x.astype(int)
```

```python
acceptreject(100)
```
